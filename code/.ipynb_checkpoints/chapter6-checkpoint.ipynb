## SGD

```python
class SGD:
    def __init__(self, lr = 0.01):
        self.lr = lr
    
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key]
```

## Momentum

```python
class Momentum:
    def __init__(self, lr = 0.01, momentum = 0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None
    
    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items():
                self.v[key] = np.zeros_like(val)
        
        for key in params.keys():
            self.v[key] = self.momentum*self.v[key] - self.lr *grads[key]
            params[key] += self.v[key]
```

### AdaGrad

```python
class AdaGrad:
    def __init__(self, lr = 0.01):
        self.lr = lr
        self.h = None
        
    def update(self, params, grads):
        # initialize h (squared sum of gradient)
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
                
        # update
        for key in params.keys():
            self.h[key] += grads[key] *grads[key]
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key] + 1e-7))
```

## array and dict

```python
def to_np_array(as_list):
    return np.array([as_list['x'], as_list['y']])

def to_dict(vector):
    as_list = {}
    as_list['x'] = vector[0]
    as_list['y'] = vector[1]
    return as_list
```

## Main

```python
def main(updater = SGD, lr = 0.99, iters = 100):
    X,Y = np.meshgrid(np.arange(-10.0,10.0, 1.0), np.arange(-5.0,5.0,0.5))
    updater = updater(lr = lr)
    
    point = np.array([-5.0,1.0])
    params = to_dict(point)

    trajectory = list() 
    trajectory.append(params)
    grad_movement = list()
    
    
    for i in range(iters):
        # get gradient
        # all the other functions are made to take dictionaries as the input
        # however, for simplicity, the numerical gradient in this file takes arrays as an input
        # therefore, we need to put in functions to put in dict -> array to get the gradient
        # and array -> dict to keep track of the trajectory
        param = to_np_array(params)
        grad = numerical_gradient(long_tube, param)

        # make into a dict 
        params = to_dict(param)
        grads = to_dict(grad)

        # update the point as lr *grad
        updater.update(params, grads)

        # add point
        trajectory.append(params)
        grad_movement.append(grads)
        
    trajectory_data = pd.DataFrame(trajectory)
    df = trajectory_data.loc[abs(trajectory_data.y) < 5, :]
    df.shape
    
    plt.contour(X, Y, long_tube([X,Y]), levels = [0.1, 0.2, 0.5, 1, 2, 5, 10, 20, 50, 100])
    plt.plot(df.x, df.y, 'ro-')
```
